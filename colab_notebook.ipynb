```python
# Cell 2: Initialize NexaAPI client
# Get your FREE API key at: https://rapidapi.com/user/nexaquency

from nexaapi import NexaAPI
import requests
from IPython.display import Image, display
import io

# Set your API key
NEXAAPI_KEY = 'YOUR_RAPIDAPI_KEY'  # Get free key: https://rapidapi.com/user/nexaquency

client = NexaAPI(api_key=NEXAAPI_KEY)

print('✅ NexaAPI client ready!')
print('💰 $0.003/image | 56+ models')
print('🔗 https://nexa-api.com')
```

```python
# Cell 3: Define Sygen agent tasks using NexaAPI
# These are the building blocks of a Sygen multi-agent system

def image_generation_task(prompt: str, model: str = 'flux-schnell') -> dict:
    """
    A Sygen background task that calls NexaAPI for image generation.
    Costs only $0.003 per image — cheapest in the market.
    
    In a real Sygen agent:
        from sygen import Agent, Tool
        class ImageAgent(Agent):
            tools = [NexaImageTool()]
    """
    print(f'🎨 [IMAGE AGENT] Generating: {prompt[:50]}...')
    
    result = client.image.generate(
        model=model,
        prompt=prompt,
        width=1024,
        height=1024
    )
    
    return {
        'image_url': result.image_url,
        'prompt': prompt,
        'model': model,
        'cost': '$0.003'
    }


def tts_narration_task(text: str, voice: str = 'en-US-neural') -> dict:
    """
    A Sygen background task that converts agent output to speech.
    """
    print(f'🔊 [NARRATOR AGENT] Converting to speech: {text[:50]}...')
    
    audio = client.audio.tts(
        text=text,
        voice=voice,
        model='tts-1'
    )
    
    return {
        'audio_url': audio.audio_url,
        'text': text,
        'voice': voice
    }

print('✅ Agent tasks defined!')
print('  - image_generation_task: Creates images via NexaAPI')
print('  - tts_narration_task: Creates audio via NexaAPI TTS')
```

```python
# Cell 4: Run the multi-agent pipeline
# Orchestrate image_generation_task + tts_narration_task

import matplotlib.pyplot as plt
from PIL import Image as PILImage

topic = 'A futuristic AI-powered city at sunset, digital art style'

print('🚀 Starting Sygen + NexaAPI multi-agent pipeline...')
print('=' * 50)

# Agent 1: Content Creator generates image
image_result = image_generation_task(
    prompt=topic,
    model='flux-schnell'
)
print(f'✅ Image URL: {image_result["image_url"][:60]}...')

# Agent 2: Narrator creates audio description
narration_text = f'Presenting: {topic}. Generated with NexaAPI at nexa-api.com'
audio_result = tts_narration_task(text=narration_text)
print(f'✅ Audio URL: {audio_result["audio_url"][:60]}...')

print('\n🎉 Pipeline complete!')
print(f'💰 Total cost: {image_result["cost"]} (image only)')
```

```python
# Cell 5: Display results

# Download and display the generated image
img_response = requests.get(image_result['image_url'], timeout=30)
img = PILImage.open(io.BytesIO(img_response.content))

plt.figure(figsize=(12, 8))
plt.imshow(img)
plt.axis('off')
plt.title(
    f'Sygen Agent Output — NexaAPI ({image_result["model"]}) — Cost: {image_result["cost"]}',
    fontsize=14, pad=15
)
plt.tight_layout()
plt.show()

print('\n📊 Results Summary:')
print(f'  🖼️  Image: {image_result["image_url"]}')
print(f'  🔊 Audio: {audio_result["audio_url"]}')
print(f'  💰 Cost: {image_result["cost"]}')
print()
print('🔗 Resources:')
print('  - NexaAPI: https://nexa-api.com')
print('  - Free key: https://rapidapi.com/user/nexaquency')
print('  - pip install nexaapi | https://pypi.org/project/nexaapi/')
print('  - npm install nexaapi | https://www.npmjs.com/package/nexaapi')
print('  - Sygen: https://pypi.org/project/sygen/')
```
